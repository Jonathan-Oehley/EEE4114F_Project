# Setup Code

```
import numpy as np
import matplotlib.pyplot as plt

from google.colab import drive
drive.mount('/content/gdrive')

model_path = "gdrive/My Drive/SVR_CSVs/" # Change this depending on what model you want to run
```

# Plot histograms

Import the model's error csv file.

```
import os
csv_arrays = []
name_counter = 1
filename = "SVR_Model_{}.csv".format(name_counter)
while os.path.isfile(model_path + filename):
    csv_arrays.append(np.genfromtxt(model_path + filename, delimiter=','))  
    name_counter += 1
    filename = "SVR_Model_{}.csv".format(name_counter)
    
print("CSV Data Imported!")
```

Palm vs Fingers

```
import matplotlib.pyplot as plt

# CSV Structure
# FrameID, Hand, Positions, Errors
#    0       1      2:19     20:

name_counter = 1
filename = "SVR_Model_{}.csv".format(name_counter)
while os.path.isfile(model_path + filename):

    #error for each frame
    mean_error_palm    = csv_arrays[name_counter-1][:, 20:22].flatten()
    mean_error_fingers = csv_arrays[name_counter-1][:, 23:].flatten()


    plt.figure(dpi=150)

    # Plotting Parameters
    a = 0.5
    x_range = None
    edge = 'k'

    plt.hist(mean_error_palm,       60, range=x_range, density=True, edgecolor=edge, alpha=a, color='b', label="Palm");
    plt.hist(mean_error_fingers,    60, range=x_range, density=True, edgecolor=edge, alpha=a, color='r', label="Fingers");

    ax = plt.gca()
    ax.set_facecolor((0.9, 0.9, 0.9))
    ax.set_title("Mean Error Distributions")
    ax.set_xlabel("Error [mm]")
    ax.set_ylabel("Probability Density")

    plt.legend(fontsize=6)
    plt.tight_layout()
    plt.savefig(model_path + "Palm_vs_Fingers_SVR_Model_{}.png".format(name_counter))
    plt.clf()
    
    name_counter += 1
    filename = "SVR_Model_{}.csv".format(name_counter)
```

Overall Mean Error

```
# CSV Structure
# FrameID, Hand, Positions, Errors
#    0       1      2:19     20:
name_counter = 1
filename = "SVR_Model_{}.csv".format(name_counter)
while os.path.isfile(model_path + filename):

    #error for each frame
    mean_error = csv_arrays[name_counter-1][:, 20:].flatten()

    plt.figure(dpi=150)

    # Plotting Parameters
    a = 0.5
    x_range = None
    edge = 'k'

    plt.hist(mean_error,       60, range=x_range, density=True, edgecolor=edge, alpha=a, color='b');

    ax = plt.gca()
    ax.set_facecolor((0.9, 0.9, 0.9))
    ax.set_title("Mean Error Distribution")
    ax.set_xlabel("Error [mm]")
    ax.set_ylabel("Probability Density")

    plt.legend(fontsize=6);



    plt.tight_layout()
    plt.savefig(model_path + "Overall_Error_SVR_Model_{}.png".format(name_counter))
    plt.clf()
    
    name_counter += 1
    filename = "SVR_Model_{}.csv".format(name_counter)
```

X, Y, Z Error

```
# CSV Structure
# FrameID, Hand, Positions, Errors
#    0       1      2:19     20:

name_counter = 1
filename = "SVR_Model_{}.csv".format(name_counter)
while os.path.isfile(model_path + filename):
      #error for each frame
      mean_error_X   = csv_arrays[name_counter-1][:, 20:-1:3].flatten()
      mean_error_Y   = csv_arrays[name_counter-1][:, 21:-1:3].flatten()
      mean_error_Z   = csv_arrays[name_counter-1][:, 22:-1:3].flatten()

      print(np.shape(mean_error_X))


      plt.figure(dpi=150)

      # Plotting Parameters
      a = 0.5
      x_range = None
      edge = 'k'

      plt.hist(mean_error_X,      60, range=x_range, density=True, edgecolor=edge, alpha=a, color='b', label="X-Error");
      plt.hist(mean_error_Y,      60, range=x_range, density=True, edgecolor=edge, alpha=a, color='r', label="Y-Error");
      plt.hist(mean_error_Z,      60, range=x_range, density=True, edgecolor=edge, alpha=a, color='w', label="Z-Error");

      ax = plt.gca()
      ax.set_facecolor((0.9, 0.9, 0.9))
      ax.set_title("Mean Error Distributions")
      ax.set_xlabel("Error [mm]")
      ax.set_ylabel("Probability Density")

      plt.legend(fontsize=6);      
      plt.tight_layout()
      plt.savefig(model_path + "XYZ_Error_SVR_Model_{}.png".format(name_counter))
      plt.clf()
      

    
      name_counter += 1
      filename = "SVR_Model_{}.csv".format(name_counter)
```
